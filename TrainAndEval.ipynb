```python
import torch
from dense_scripts.utils import DenseLunarLander
from dense_scripts.utils.policies import SimpleGRPOPolicy
from dense_scripts.PPO.ppo import PPOTrainer, PPOConfig

# 1. Initialize environments (can be env objects)
# Using the same custom envs as your GRPO example
env_train = DenseLunarLander(randomize_angle=True, gravity=-5)
env_eval = DenseLunarLander(randomize_angle=True, gravity=-5)

# 2. Initialize the Policy (Actor)
# The PPOTrainer will automatically create its own ValueNet (Critic)
policy = SimpleGRPOPolicy(env_train.observation_space.shape[0], env_train.action_space.n)

# 3. Configure the Trainer
# Pass the env_train object directly
cfg = PPOConfig(
    env=env_train,
    G=32,                 # Number of parallel workers
    T=1024,               # Steps per worker per iteration
    gamma=0.99,           # Using the gamma from your GRPO example
    log_dir="./runs/PPO_dense_eval"
)

# 4. Initialize the Trainer
# We pass the actor policy; the trainer builds the critic
trainer = PPOTrainer(policy, cfg, device="cpu")

# 5. Run Training
# The train() signature now matches your GRPO trainer
trainer.train(
    iters=300,
    eval_env=env_eval,
    eval_interval=30,
    eval_episodes=100,
    video_dir="videos/PPO",
    video_fps=30,
    video_format="mp4",
    video_episodes=12
)
```
